```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    "Product": ["Shirt", "Jeans", "Shoes", "Bag", "Cap"],
    "Price": [1200, 2200, 3500, 1800, 500],
    "Quantity_Sold": [50, 30, 20, 25, 60],
    "Quantity_Bought": [40, 25, 15, 20, 55]
}

df = pd.DataFrame(data)
```

```python
df["Revenue"] = df["Price"] * df["Quantity_Sold"]
df["Profit"] = df["Revenue"] - (df["Price"] * df["Quantity_Bought"])
```

```python
print("ðŸ“Š Sales & Orders Data Overview:\n")
print(df)
```

```python
print("\nðŸ”Ž Summary Statistics:\n")
print(df.describe())
```

```python
plt.figure(figsize=(8,5))
plt.bar(df["Product"], df["Revenue"], color="skyblue")
plt.title("Revenue by Product")
plt.xlabel("Products")
plt.ylabel("Revenue (NPR)")
plt.show()
```
